# Modelling

## 1. Vectorizing Raw Data

### 1.1 Count Vectorization

```python
# Initiate the stopwords and lemmatization
stopwords = nltk.corpus.stopwords.words('english')
wn = nltk.WordNetLemmatizer()
```

```python
# Defining a function to do all followings: Remove punctuations, tokenize, remove stopwords, and lemmatize
def clean_text(text):
    # Remove unicode characters
    text = text.encode('ascii', 'ignore').decode()
    
    # Remove punctuation and lowercase the words
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    
    # \W matches any non-word character (equivalent to [^a-zA-Z0-9_]). This does not include spaces i.e. \s
    # Add a + just in case there are 2 or more spaces between certain words
    tokens = re.split('\W+', text)
    
    # Apply lemmatization and stopwords exclusion within the same step
    text = [wn.lemmatize(word) for word in tokens if word not in stopwords]
    return text
```

```python
# Create an instance of CountVectorizer and pass in the clean_text function (defined above) as the analyzer parameter, extracting max_features of 10,000
count_vect = CountVectorizer(analyzer=clean_text, max_features = 10000)

# Stores the vectorized version of the data in title_text_counts
title_text_counts = count_vect.fit_transform(train['title_text'])

# Returns the number of rows and columns of title_text_counts i.e. 19510 entries and 10000 unique words
print(title_text_counts.shape)
```

```python
# Expand 'title_text_counts' out to a collection of arrays and then store it in a data frame
title_text_counts_df = pd.DataFrame(title_text_counts.toarray())

# Rename the columns with the words
title_text_counts_df.columns = count_vect.get_feature_names()

# Print a few rows of'title_text_counts_df'
title_text_counts_df.head()
```

### 1.2 TFIDF Vectorization

```python
# Create an instance of TfidfVectorizer and pass in the clean_text function (defined above) as the analyzer parameter, extracting max_features of 10,000
tfidf_vect = TfidfVectorizer(analyzer=clean_text, max_features = 10000)

# Stores the vectorized version of the data in title_text_tfidf
title_text_tfidf = tfidf_vect.fit_transform(train['title_text'])

# Returns the number of rows and columns of title_text_tfidf i.e. 19510 entries and 10000 unique words
print(title_text_tfidf.shape)
```

```python
# Expand 'title_text_tfidf' out to a collection of arrays and then store it in a data frame
title_text_tfidf_df = pd.DataFrame(title_text_tfidf.toarray())

# Rename the columns with the words
title_text_tfidf_df.columns = tfidf_vect.get_feature_names()

# Print a few rows of'title_text_tfidf_df'
title_text_tfidf_df.head()
```

### 3.2 Logistics Regression

### 3.2.1 Logistics Regression Using Count Vectorization

```python
# Run start time in seconds
start = time()

# Instantiate logistic regression in pipeline
pipe_lr = Pipeline([
    ('lr', LogisticRegression(max_iter=5))])

# A list of params to be fit into RandomizedSearch CV
pipe_lr_params = {'lr__solver': ['lbfgs', 'newton-cg', 'saga', 'liblinear'], 
             'lr__penalty': [None,'l1','l2'], 
             'lr__C': np.linspace(0.01, 1, 5)
            }

# Instantiate RandomizedSearchCV
randsearch_lr = RandomizedSearchCV(pipe_lr, pipe_lr_params)

# Fit the data
randsearch_lr.fit(X_count_train, y_count_train)

# The best params to used for the model
randsearch_lr_best = randsearch_lr.best_params_
print('Best Parameters:', randsearch_lr_best)

# Amount of time taken to run in seconds
time_taken = time() - start
print('Time taken to run:', round(time_taken, 2), 'seconds')
```

The optimal hyperparameters values for the Logistic Regression Model found are:
- solver = liblinear
- C = 0.2575
- penalty = l1

We will run the model once again with this value and pull out the scores required for evaluation.

```python
# Initiate the logistic Regression Model
lr_optimal = LogisticRegression(max_iter=100, solver='liblinear', C=0.2575, penalty='l1')

# Fit data into the Model
lr_optimal.fit(X_count_train, y_count_train)

# Print the model name to ensure that it works.
print(lr_optimal)

# Use model to predict classification
lr_y_pred = lr_optimal.predict(X_count_test)

# Get mean Train Accuracy using cross-Validation with number of folds = 5
lr_train_accu = cross_val_score(lr_optimal, X_count_train, y_count_train, cv=5).mean()
print('\nLogistic Regression Train Accuracy:', lr_train_accu)

# Get mean Test Accuracy using cross-Validation with number of folds = 5
lr_test_accu = cross_val_score(lr_optimal, X_count_test, y_count_test, cv=5).mean()
print('Logistic Regression Test Accuracy:', lr_test_accu)

# Print classification report for y_pred compared to y_test
lr_report = classification_report(y_count_test, lr_y_pred)
print('\nLogistic Regression Classification Report:\n', lr_report)

# Print roc_auc_score for y_test
lr_auc = roc_auc_score(y_count_test, lr_optimal.predict_proba(X_count_test)[:,1])
print('Logistic Regression AUC value:', lr_auc)
```

### 3.9 BERT

#### Prepare Data

```python
# Create features matrix (X) and target vector (y)
X_bert = train['title_text_clean']    
y_bert = train['label']
```

```python
# Train-test split is done with 70/30 split and stratifying the y
X_bert_train, X_bert_test, y_bert_train, y_bert_test = train_test_split(X_bert, y_bert, test_size=.3, stratify=y_bert)
```

```python
# Save the X_bert_train and y_bert_train into a df
bert_train_df = pd.DataFrame([X_bert_train,y_bert_train]).T.reset_index()

# Save the X_bert_test and y_bert_test into a df
bert_test_df = pd.DataFrame([X_bert_test,y_bert_test]).T.reset_index()
```

```python
# Generate a list of dictionaries
# Initiate a new list for train
bert_train_list_dict = []

# Append the dictionary items of train dataset into the list, key: title_text_clean, values: label
for i in range (0,len(bert_train_df['title_text_clean'])):
    bert_train_list_dict.append({'text':bert_train_df['title_text_clean'][i], 'label':bert_train_df['label'][i]})
```

```python
# Generate a list of dictionaries
# Initiate a new list for test
bert_test_list_dict = []

# Append the dictionary items of test dataset into the list, key: title_text_clean, values: label
for i in range (0,len(bert_test_df['title_text_clean'])):
    bert_test_list_dict.append({'text':bert_test_df['title_text_clean'][i], 'label':bert_test_df['label'][i]})
```

```python
# Generate a list of tuples from the list above from train dataset
bert_train_texts, bert_train_labels = list(zip(*map(lambda d: (d['text'], d['label']), bert_train_list_dict)))
```

```python
# Generate a list of tuples from the list above from test dataset
bert_test_texts, bert_test_labels = list(zip(*map(lambda d: (d['text'], d['label']), bert_test_list_dict)))
```

```python
# Generate tokens and token ids of the train and test texts
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)

train_tokens = list(map(lambda t: ['[CLS]'] + tokenizer.tokenize(t)[:511], bert_train_texts))
test_tokens = list(map(lambda t: ['[CLS]'] + tokenizer.tokenize(t)[:511], bert_test_texts))

train_tokens_ids = list(map(tokenizer.convert_tokens_to_ids, train_tokens))
test_tokens_ids = list(map(tokenizer.convert_tokens_to_ids, test_tokens))

train_tokens_ids = pad_sequences(train_tokens_ids, maxlen=50, truncating="post", padding="post", dtype="int")
test_tokens_ids = pad_sequences(test_tokens_ids, maxlen=50, truncating="post", padding="post", dtype="int")
```

```python
# Generate a boolean array based on value of 'label' for train and test sets
train_y = np.array(bert_train_labels) == 1
test_y = np.array(bert_test_labels) == 1
```

#### Model Building and Predictions

```python
# Create a BERT classifier which contains an ‘initialization’ method and a ‘forward’ method that returns token probabilities
class BertBinaryClassifier(nn.Module):
    def __init__(self, dropout=0.1):
        super(BertBinaryClassifier, self).__init__()
        
        self.bert = BertModel.from_pretrained('bert-base-uncased')

        self.dropout = nn.Dropout(dropout)
        self.linear = nn.Linear(768, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, tokens, masks=None):
        _, pooled_output = self.bert(tokens, attention_mask=masks, output_all_encoded_layers=False)
        dropout_output = self.dropout(pooled_output)
        linear_output = self.linear(dropout_output)
        proba = self.sigmoid(linear_output)
        return proba
```

```python
# Generate train and test masks
train_masks = [[float(i > 0) for i in ii] for ii in train_tokens_ids]
test_masks = [[float(i > 0) for i in ii] for ii in test_tokens_ids]

train_masks_tensor = torch.tensor(train_masks)
test_masks_tensor = torch.tensor(test_masks)
```

```python
# Generate token tensors for training and testing
train_tokens_tensor = torch.tensor(train_tokens_ids).to(torch.int64)
train_y_tensor = torch.tensor(train_y.reshape(-1, 1)).to(torch.int64).float()

test_tokens_tensor = torch.tensor(test_tokens_ids).to(torch.int64)
test_y_tensor = torch.tensor(test_y.reshape(-1, 1)).to(torch.int64).float()
```

```python
# Prepare data loaders
train_dataset =  torch.utils.data.TensorDataset(train_tokens_tensor, train_masks_tensor, train_y_tensor)
train_sampler =  torch.utils.data.RandomSampler(train_dataset)
train_dataloader =  torch.utils.data.DataLoader(train_dataset, sampler=train_sampler)
test_dataset =  torch.utils.data.TensorDataset(test_tokens_tensor, test_masks_tensor, test_y_tensor)
test_sampler =  torch.utils.data.SequentialSampler(test_dataset)
test_dataloader =  torch.utils.data.DataLoader(test_dataset, sampler=test_sampler)
```

```python
# Fine tuning the BERT model by using the Adam optimizer to minimise the Binary Cross Entropy loss
EPOCHS = 2
bert_clf = BertBinaryClassifier()
optimizer = torch.optim.Adam(bert_clf.parameters(), lr=2e-6)
for epoch_num in range(EPOCHS):
    bert_clf.train()
    train_loss = 0
    for step_num, batch_data in enumerate(train_dataloader):
        token_ids, masks, labels = tuple(t for t in batch_data)
        probas = bert_clf(token_ids, masks)
        loss_func = nn.BCELoss()
        batch_loss = loss_func(probas, labels)
        train_loss += batch_loss.item()
        bert_clf.zero_grad()
        batch_loss.backward()
        optimizer.step()
        print('Epoch: ', epoch_num + 1)
        print("\r" + "{0}/{1} loss: {2} ".format(step_num, len(train_dataloader), train_loss / (step_num + 1)))
```

```python
# Evaluate the BERT Model
bert_clf.eval()
bert_predicted = []
all_logits = []
with torch.no_grad():
    for step_num, batch_data in enumerate(test_dataloader):
        token_ids, masks, labels = tuple(t for t in batch_data)
        logits = bert_clf(token_ids, masks)
        loss_func = nn.BCELoss()
        loss = loss_func(logits, labels)
        numpy_logits = logits.cpu().detach().numpy()
        
        bert_predicted += list(numpy_logits[:, 0] > 0.5)
        all_logits += list(numpy_logits[:, 0])
```

```python
# Print classification report for bert_predicted compared to test_y
bert_report = classification_report(test_y, bert_predicted)
print('\nBERT Classification Report:\n', bert_report)

# Print roc_auc_score for y_test
bert_auc = roc_auc_score(test_y, bert_predicted)
print('BERT AUC value:',bert_auc)
```

## 4. Model Evaluation

Recall that the news articles are classified as such:

|News Articles Classification|Numerical Value in Dataframes|Class|
|:---:|:---:|:---:|
|**Fake News Article**|1|Positive|
|**Real News Article**|0|Negative|

The metrics that we will use to decide the best model are:
- Test Accuracy
- Train Accuracy - Test Accuracy (to check for overfitting)
- Precision
- Recall
- F1 score
- Area Under Curve (AUC)

#### *Test Accuracy*

$$Accuracy = \frac{\text{True Positives + True Negatives}}{\text{Total Predictions}}$$

When we analysed our Baseline Model, we noted that the Accuracy score is a good metric to use when evaluating a balanced datasets like the one used in this project.

The Accuracy score gives an overview of how accurate the model is in predicting both the majority and the minority classes combined. 

**The higher the test accuracy score (as close to 1 as possible), the better is the prediction of the model.**

#### *Train Accuracy - Test Accuracy*

To check if a model is overfitted or underfitted, we will need to compare the Train Accuracy with the Test Accuracy within each model. 

If Train-Test Accuracy gives a positive number (>0), this indicates that the model is overfitting. The more positive is the number, the more serious is the overfitting. Practically, if the difference between train accuracy and test accuracy is within 0.1, it is still acceptable.

If Train-Test Accuracy gives a negative number (<0), this indicates that the model is underfitting. The more negative is the number, the more serious is the underfitting. Practically, if the difference between train accuracy and test accuracy is within 0.1, it is still acceptable.

**Ideally, we would like to see train - test accuracy to be as close to 0 as possible. If not, the difference within 0.1 is still acceptable.**

#### *Precision score*

$$Precision = \frac{\text{True Positives}}{\text{True Positives + False Positives}}$$

Precision is the ratio of True Positives to All Predicted Positives. It means that of all the Positive predictions made by the model, what percentage of this is actually correct.

Put simply in the context of this project, the higher the Precision score, the higher the number of news articles that are predicted as fake actually are indeed fake.

**The higher the precision score (as close to 1 as possible), the better is the prediction of model.**

#### *Recall score*

$$Recall = \frac{\text{True Positives}}{\text{True Positives + False Negatives}}$$

Recall is the ratio of True Positives to All Actual Positives. It shows the percentage of correctly classified Positive predictions by the model against all the positives in the dataset.

Put simply in the context of this project, the higher the Recall score, the higher the number of news article is correctly classified as fake.

**The higher the recall score (as close to 1 as possible), the better is the prediction of model.**

#### *F1 score*

$$F1 score  =  2 * \frac{\text{Precision * Recall}}{\text{Precision + Recall}}$$

The F1 score is the weighted average of the Precision-Recall scores. It will be used to give a better perspective of which model performs the best. 

When there are 2 or more models with similar Precision-Recall scores and a clear answer of which model is better cannot be derived just by comparing the Precision-Recall scores themselves, we will use the F1 score to help us to decide on which model to choose.

An F1 score of 0 means the model is as bad as it can get and a score of 1 means the model is perfect. 

**The higher the F1 Score (as close to 1 as possible), the better is the prediction of model.**

#### *Area Under Curve (AUC)*

AUC is the area under the Receiver Operating Characteristics (ROC) curve, with False Positive Rate (FPR) on the x-axis and True Positive Rate (TPR) on the y-axis. The AUC score will give us an overall measure of the performance of our classifier. The AUC value lies between 0.5 to 1 where 0.5 denotes a bad classifer and 1 denotes an excellent classifier.

A high AUC value indicates that our model is generally good at distinguishing between the Positive (Fake News Articles) and Negative (Real New Articles) classes. 

**The higher AUC (as close to 1 as possible), the better is the prediction of model.**

In evaluating our models, the priority of metrics is as follows:
1. Test Accuracy
2. F1 score
3. AUC
4. Train-Test Accuracy


Firstly, we look at Test Accuracy, as the test accuracy gives us the indicator on how well the prediction is done by the model. A higher accuracy score means the prediction is more accurate, less articles being wrongly classified.

Secondly, we look at the F1 score, which gives us a balance perspective on the precision-recall score performance. A higher F1 score means the combination of precision-recall of the models gives the better performance.

Next, we look at AUC score, as it gives us the overall measure of performance of our classifier. By analogy, the higher the AUC, the better the model is at distinguishing between real news articles and fake news articles.

Lastly, we look at the difference between train accuracy and test accuracy. 
If train - test accuracy > 0, there is overfitting occurs.
If train - test accuracy < 0, there is underfitting occurs.

It's the most optimum when the train - test accuracy is as close to 0 as possible, indicating the overfitting/underfitting is less severe.

**Based on the above metrics, XGBoost model using Count Vectorization is chosen.**

This is because XGBoost model using Count Vectorization have the highest test accuracy score out of all models, as well as having the highest F1 score out of all models, and the highest AUC score out of all models. With regards to the difference between train accuracy and test accuracy, XGBoost model using Count Vectorization have train - test accuracy > 0, hence it is overfitting. However, the difference is minimal (~0.01), hence the overfitting is not that severe.
